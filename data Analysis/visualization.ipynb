Graph types
- line graph : show trends over time
- Bar Graph : compare values between categories (discrete data)
- pie chart : compare values between categories (percentage data)
- Histogram : show distribution of numerical data
- Box Plot : statstical distribution of numeric data
- Heat Map : Show matrix of values using color
- Scatter Plot : Show relationship between two variables
- Bubble Chart : Show relationship between three variable 

```python
import seaborn as sns 
```

```python
tips =  sns.load_dataset("tips")
tips
```

```python

plt.plot('total_bill', data=tips)
```

```python
fig, ax = plt.subplots(figsize = (12,4))
plt.plot('total_bill', data=tips)
```

```python
fig, ax = plt.subplots(figsize = (12,4))
ax.plot('total_bill', data=tips, color='red', marker =".")
```

```python
fig, ax= plt.subplots(figsize=(20,4))
ax.bar(tips.index, 'total_bill', data=tips, color='red')
ax.bar(tips.index, 'tip', data=tips, color='#99ff99')
ax.set_xlim(0,100)
```

```python
plt.figure(figsize=(20,4))
plt.stem('total_bill',data= tips)
```

```python
#histogram
plt.hist('total_bill', data = tips, bins=range (0,60,5))
plt.show()
```

```python
 #histogram
h = plt.hist ('total_bill' , data=tips, bins=range(0,60,5),
              color = 'green', edgecolor='black', alpha=.5, hatch ='//')
for i in range (len(h[0])):
    x = h[1] [i] + 1.5
    y = h[0] [i] + 2
    s = str(int(h[0][i]))
    plt.text(x,y,s)
plt.ylim(0.80)
plt.show()    
```

```python
gender_count = tips.sex.value_counts()
gender_count
```

```python
plt.pie(gender_count.values,
        labels=gender_count.index,
        autopct='%.1f%%')
plt.show()
```

```python
#donut
plt.pie(gender_count.values,
        labels= gender_count.index,
        autopct='%.1f%%',
        wedgeprops={'width':.6},
        pctdistance=.65,
        colors=['#ff9999', '#66b3ff'])
plt.show()
```

pie chart should not be used with alot of different categories

```python
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt. subplots(2,3, figsize = (20,10))
gender_count = tips.sex.value_counts()
ax1.pie(gender_count.values,
        labels = gender_count.index,
        autopct = '%.1f%%',
        wedgeprops={'width':.6},
        pctdistance=.65,
        colors=['#ff9999', '#66b3ff'])

smoker_count = tips.smoker.value_counts()
ax2.pie(smoker_count.values,
        labels=smoker_count.index,
        autopct='%.1f%%', 
        wedgeprops={'width':.6},
        pctdistance=.65,
        colors=['#ff9999','#66b3ff']) 

day_count = tips.day.value_counts()
ax3.pie(day_count.values,
        labels = day_count.index,
        autopct = '%.1f%%',
        wedgeprops={'width':.6},
        pctdistance=.65,
        colors=['#ff9999', '#66b3ff'])

time_count = tips.time.value_counts()
ax4.pie(time_count.values,
        labels = time_count.index,
        autopct = '%.1f%%',
        wedgeprops={'width':.6},
        pctdistance=.65,
        colors=['#ff9999', '#66b3ff'])

size_count = tips['size'].value_counts()
ax5.pie(size_count.values,
        labels = size_count.index,
        autopct = '%.1f%%',
        wedgeprops={'width':.6},
        pctdistance=.65,
        colors=['#ff9999', '#66b3ff'])
ax6.axis('off')
plt.show()

```
